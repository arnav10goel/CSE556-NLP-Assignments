```python
import json
from sentence_transformers import SentenceTransformer
import pickle

# Load the model
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Import the json file
with open('train.json') as f:
    train_data = json.load(f)
```

```python
# Store the embeddings:
train_embeddings_dict = {}

i = 0
for train_dict in train_data:
  # Get the relevant information from the dictionary
  episode_key = train_dict["episode"]
  utterances = train_dict["utterances"]

  # Generate embeddings for the utterances
  embeddings = model.encode(utterances)
  i+=1
  # Print
  print(embeddings.shape)
  print(i)

  # Store the embedding
  train_embeddings_dict[episode_key] = embeddings
```

```python
# Store these embeddings in a pickle file
with open('train_embeddings_.pkl', 'wb') as f:
    pickle.dump(train_embeddings_dict, f)
```
