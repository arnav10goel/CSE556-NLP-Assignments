```python
import json
from collections import defaultdict

def assign_bio_labels(tokens, aspect_tags):
    bio_labels = ['O'] * len(tokens)
    for aspect in aspect_tags:
        start=aspect['from']
        end=aspect['to']
        bio_labels[start]='B'
        for i in range(start+1,end):
            bio_labels[i]='I'
    return bio_labels


#####################################################
# Load the train dataset
with open('Laptop_Review_Train.json', 'r', encoding='utf-8') as file:
    train_data = json.load(file)

# Initialize a dictionary to store the train output
train_output = defaultdict(dict)

num_reviews = 1
# Process each data instance in the train dataset
for instance in train_data:
    text=instance['raw_words']
    tokens=instance['words']
    aspect_tags=instance['aspects']
    bio_label=assign_bio_labels(tokens, aspect_tags)

    train_output[num_reviews]['text'] = text
    train_output[num_reviews]['labels'] = bio_label

    num_reviews += 1

# Save the train output to a file
with open('Laptop_Review_Train_BioLabels.json', 'w', encoding='utf-8') as file:
    json.dump(train_output, file, ensure_ascii=False, indent=4)

#####################################################
# Load the validation dataset
with open('Laptop_Review_Val.json', 'r', encoding='utf-8') as file:
    validation_data = json.load(file)

# Initialize a dictionary to store the validation output
validation_output = defaultdict(dict)

num_reviews = 1
# Process each data instance in the validation dataset
for instance in validation_data:
    text=instance['raw_words']
    tokens=instance['words']
    aspect_tags=instance['aspects']
    bio_label=assign_bio_labels(tokens, aspect_tags)

    validation_output[num_reviews]['text'] = text
    validation_output[num_reviews]['labels'] = bio_label

    num_reviews += 1

# Save the validation output to a file
with open('Laptop_Review_Validation_BioLabels.json', 'w', encoding='utf-8') as file:
    json.dump(validation_output, file, ensure_ascii=False, indent=4)

#####################################################
# Load the test dataset
with open('Laptop_Review_Test.json', 'r', encoding='utf-8') as file:
    test_data = json.load(file)

# Initialize a dictionary to store the test output
test_output = defaultdict(dict)

num_reviews = 1
# Process each data instance in the test dataset
for instance in test_data:
    text=instance['raw_words']
    tokens=instance['words']
    aspect_tags=instance['aspects']
    bio_label=assign_bio_labels(tokens, aspect_tags)

    test_output[num_reviews]['text'] = text
    test_output[num_reviews]['labels'] = bio_label

    num_reviews += 1

# Save the test output to a file
with open('Laptop_Review_Test_BioLabels.json', 'w', encoding='utf-8') as file:
    json.dump(test_output, file, ensure_ascii=False, indent=4)

```
